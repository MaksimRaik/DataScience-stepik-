```python
import numpy as np
import pandas as pd
```

```python

student_perfomance  = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

Рассчитать среднее значение оценок студентов в зависимости от пола

```python
student_perfomance.groupby('gender').aggregate({'math score': 'mean', 'reading score': 'mean', 'writing score': 'mean'})
```

Группировка по нескольким переменным

```python
student_perfomance.groupby(['gender', 'race/ethnicity']).aggregate({'math score': 'mean', 'reading score': 'mean', 'writing score': 'mean'})
```

Сохранить успеваемость студентов

```python
student_perfomance['total_score'] = student_perfomance['math score'] + student_perfomance['reading score']+ student_perfomance['writing score']
```

```python
student_perfomance = student_perfomance.assign(total_score_log = np.log(student_perfomance.total_score))
```
