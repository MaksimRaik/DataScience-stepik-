```python
df = pd.read_csv('https://stepik.org/media/attachments/course/4852/dota_hero_stats.csv')
```

```python
table = df.groupby(['attack_type', 'primary_attr']).aggregate({'id':'count'})
table.sort_values('id', ascending=False)
```
