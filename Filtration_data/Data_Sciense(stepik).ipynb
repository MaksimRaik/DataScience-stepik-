```python
import numpy as np
import pandas as pd
```

```python
student_perfomance  = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
student_perfomance.gender == 'female'
```

```python
student_perfomance.loc[student_perfomance.gender == 'female']
```

Отобрать конкретные столбцы из предыдущего запроса

```python
student_perfomance.loc[student_perfomance.gender == 'female', ['gender','writing score']]
```

Отобрать только те строки, где *writing score* превышает среднее значение в данном датафрейме.

```python
AVG_writing_score = student_perfomance['writing score'].mean()
```

```python
student_perfomance.loc[student_perfomance['writing score'] > AVG_writing_score , ['gender','writing score']]
```

Отобрать только те строки, где *writing score* превышает среднее значение в данном датафрейме и пол женский.

```python
student_perfomance.loc[(student_perfomance['writing score'] > AVG_writing_score) & (student_perfomance.gender == 'female'), ['gender','writing score']]
```

У какой доли студентов из датасэта в колонке lunch указано free/reduced?

Формат ответа десятичная дробь, например, 0.25

```python
student_perfomance.loc[student_perfomance.lunch == 'free/reduced', ['lunch']].size / student_perfomance.lunch.size
```

Как различается среднее и дисперсия оценок по предметам у групп студентов со стандартным или урезанным ланчем?

*aggregate* - объединяет данные https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.aggregate.html

```python
student_perfomance.groupby('lunch').aggregate({'math score':['mean', 'var'], 'reading score':['mean', 'var'], 'writing score' :['mean', 'var']})
```

Переименования заголовков

```python
student_perfomance = student_perfomance.rename( columns = {'parental level of education': 'parental_level_of_education',
                                                           'test preparation course': 'test_preparation_course',
                                                           'math score': 'math_score',
                                                           'reading score': 'reading_score',
                                                           'writing score': 'writing_score'} )
```

Метод *query()*

```python
a = 75

student_perfomance.query(f"writing_score > {a} & gender == 'female'")
```

```python
a = 75

student_perfomance.query("writing_score > @a & gender == 'female'")
```

Отобрать студентов, уровень образования родителей которых бакалавр или магистр

```python
student_perfomance[student_perfomance['parental_level_of_education'].isin(["bachelor's degree", "master's degree"])]
```

Отобрать студентов, которые не прошли тест

```python
student_perfomance[student_perfomance['test_preparation_course'] != 'completed']
```

Отобрать студентов, оценка за математику которых выше 90 и за чтение меньше либо равна 72

```python
student_perfomance[(student_perfomance['math_score'] > 90) & (student_perfomance['reading_score'] <=72)]
```

Отобрать студентов со стандартным ланчем

```python
student_perfomance.query("lunch == 'standard'")
```
